```python
import pandas as pd
import numpy as np
```

```python
whr_2018 = './WHR_2018.csv'
happiness_df = pd.read_csv(whr_2018)

govt_expenditure = './govt_expenditure_gdp_percentage.csv'
govt_expenditure_df = pd.read_csv(govt_expenditure)
```

```python
happiness_df['country'] = happiness_df['country'].str.strip()
govt_expenditure_df['country'] = govt_expenditure_df['country'].str.strip()

union_df = happiness_df.merge(govt_expenditure_df.dropna(), on='country', how='outer')
cleaned_union_df = union_df.dropna()

cleaned_union_df = cleaned_union_df.sort_values(['average_happiness'], ascending=[True])
```

```python
cleaned_union_df = cleaned_union_df.reset_index(drop=True).reset_index()
```

```python
cleaned_union_df
```

```python
plt = cleaned_union_df.plot(
    kind='scatter', x='index', y='gov_expenditure_gdp_percentage', 
    figsize=(40,20), 
    use_index=True,
    xticks=cleaned_union_df['index']
)

# plt.set_xticks(cleaned_union_df['average_happiness'])
# plt.set_xticks(cleaned_union_df['index'])
plt.set_xticklabels(cleaned_union_df['country'], rotation='vertical')

plt.set_xlabel("Happiness report 2018")
plt.set_ylabel("Govt expenditure % GDP")
plt
```
